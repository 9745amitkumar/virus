```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
df=pd.read_csv("/content/heart.csv")
df.head()
```

```python
x=df.drop("target",axis=1)
y=df["target"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
linear=svm.SVC(kernel="linear")
sigmoid=svm.SVC(kernel="sigmoid")
rbf=svm.SVC(kernel="rbf")
poly=svm.SVC(kernel="poly")
linear.fit(x_train,y_train)
sigmoid.fit(x_train,y_train)
rbf.fit(x_train,y_train)
poly.fit(x_train,y_train)
```

```python
y_linear=linear.predict(x_test)
y_sigmoid=sigmoid.predict(x_test)
y_rbf=rbf.predict(x_test)
y_poly=poly.predict(x_test)
```

```python
print("Accuracy of linear , sigmoid , rbf, poly models are : ",accuracy_score(y_test,y_linear),accuracy_score(y_test,y_sigmoid),accuracy_score(y_test,y_rbf),accuracy_score(y_test,y_poly))
```

```python
print("confusion matrix for linear :")
print(confusion_matrix(y_test,y_linear))
print("classification report for linear model:")
print(classification_report(y_test,y_linear))
```
